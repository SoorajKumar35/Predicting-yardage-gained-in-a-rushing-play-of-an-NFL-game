Regression Models

```
def sklearn_fit_predict(model, X_train, X_val, X_test, Y_train, Y_val, Y_test):

  model.fit(X_train, Y_train)
  Y_test_pred = model.predict(X_test)

  return Y_test_pred
```

```
from sklearn import metrics
def calc_regression_error(Y_test, Y_test_pred):
  errors = {
      'mean_absolute_error': metrics.mean_absolute_error(Y_test, Y_test_pred),
      'mean_squared_error': metrics.mean_squared_error(Y_test, Y_test_pred),
  }
  return errors
```

```
# Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor

dtc = DecisionTreeRegressor(splitter='random', max_depth=10)
dtc_test_predicts = sklearn_fit_predict(dtc, X_train, X_val, X_test, Y_train, Y_val, Y_test)
dtc_regression_errors = calc_regression_error(Y_test, dtc_test_predicts)
dtc_regression_errors
```

```
# K-nearest neighbor Regressor
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor()
knn_test_predicts = sklearn_fit_predict(knn, X_train, X_val, X_test, Y_train, Y_val, Y_test)
knn_regression_errors = calc_regression_error(Y_test, knn_test_predicts)
knn_regression_errors
```

```
# Linear Regression
from sklearn.linear_model import LinearRegression as LR

lr = LR()
lr_test_predicts = sklearn_fit_predict(lr, X_train, X_val, X_test, Y_train, Y_val, Y_test)
lr_regression_errors = calc_regression_error(Y_test, lr_test_predicts)
lr_regression_errors
```

Classification Models

```
def Y_to_199_one_hot(Y):
  y = np.zeros((Y.shape[0], 199))
  for idx, target in enumerate(list(Y)):
      y[idx][99 + target] = 1
  return y
```

```
def pred_probs_to_199_classes(probs, curr_classes):
  y = np.zeros((probs.shape[0], 199))
  for i in range(probs.shape[0]):
    for class_idx, class_prob in enumerate(list(probs[i])):
        y[i][99+curr_classes[class_idx]] = class_prob
  return y
```

```
def calc_crps(Y, Y_pred):
  y_true = np.clip(np.cumsum(Y, axis=1), 0, 1)
  y_pred = np.clip(np.cumsum(Y_pred, axis=1), 0, 1)
  val_s = ((y_true - y_pred) ** 2).sum(axis=1).sum(axis=0) / (199 * Y.shape[0])
  val_s = np.round(val_s, 6)
  return val_s
```

```
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, Y_train)
nb_test_predicts = nb.predict_proba(X_test)
nb_crps_error = calc_crps(Y_to_199_one_hot(Y_test), pred_probs_to_199_classes(nb_test_predicts, nb.classes_))
print(nb_crps_error)
```

```
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=10)
dtc.fit(X_train, Y_train)
dtc_test_predicts = dtc.predict_proba(X_test)
dtc_crps_error = calc_crps(Y_to_199_one_hot(Y_test), pred_probs_to_199_classes(dtc_test_predicts, dtc.classes_))
print(dtc_crps_error)
```

```
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train, Y_train)
dtc_test_predicts = dtc.predict_proba(X_test)
dtc_crps_error = calc_crps(Y_to_199_one_hot(Y_test), pred_probs_to_199_classes(dtc_test_predicts, dtc.classes_))
print(dtc_crps_error)
```
